```python
import pandas as pd
```

***Create a DataFrame from Dictionary***

```python
starting = {'Col1': [1,2,3,4],'Col2': [3,4,5,6],'Col3': [5,6,7,8],}
df = pd.DataFrame(starting)
```

```python
print(df)
print(df.dtypes)
```

```python
starting = {'Col1': ['1','2','3','4'],'Col2': [3,4,5,6],'Col3': [5,6,7,8],}
df = pd.DataFrame(starting)
print(df)
print(df.dtypes)
```

***We reference an item through column and index***

```python
print(df['Col1'][2])
```

```python
df['Col1']
```

```python
df.Col1
```

```python
df.Col1[2]
```

***Indexing***

```python
starting = {'Col1': [1,2,3,4],'Col2': [3,4,5,6],'name':['sh', 'pr', 'dp','rp'], 'Col3': [5,6,7,8],}
df = pd.DataFrame(starting)
```

```python
df
```

```python
df = df.set_index('name')
```

```python
print(df.index)
```

```python
print(df)
```

```python
df = df.reset_index()
```
