```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
import pandas_datareader.data as web
```

```python
from matplotlib import style
style.use("fivethirtyeight")
```

```python
start = datetime.datetime(2005,1,1)
end = datetime.datetime(2015,1,1)
att = web.DataReader("T", 'yahoo', start, end)
```

```python
att.head()
```

```python
att['Open_Original'] = att['Open']
att['Open'] = att['Open']/10
```

```python
att.head()
```

```python
att['High_minus_low'] = att['High'] - att['Low']
```

```python
att.head()
```

```python
att['Daily_Raise'] = att['Close'] > att['Open_Original']
```

```python
print(att.head())
```

```python
att2 = att[(att['Close'] > att['Open_Original'])]
```

```python
att2.head()
```

```python
att['HL'] = att['High'] - att['Low']
```

```python
att2 = att[(att['HL']>1)]
```

```python
att['50MA'] = att['Close'].rolling(50 ).mean()
```

```python
att['10MA'] = att['Close'].rolling(10).mean()
```

```python
att.head(13)
```

```python
att.tail()
```

```python
att['50std'] =  att['Close'].rolling(50).std()
```

```python
att.tail()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)

att['50std'].plot(ax=axes[1], label='50STD')
att['Close'].plot(ax=axes[0], label='Price')
att['50MA'].plot(ax=axes[0], label='50MA')
att['10MA'].plot(ax=axes[0], label='10MA')
plt.legend(loc=4)
```

```python
import statistics 

def moving_average(data):
    return statistics.mean(data) 
```

```python
att['MA_With_Apply'] = att['Close'].rolling(50).apply(moving_average)
```
